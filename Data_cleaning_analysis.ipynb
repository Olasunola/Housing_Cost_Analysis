```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("full_list.csv")
```

```python
df.head()
```

```python
rough_list = []
for i in df["0"]:
    rough_list.append(i)
print(rough_list)
```

```python
refined_price = []
for p in rough_list:
    refined_price.append(p.rpartition(";")[0])
print(refined_price)
```

```python
price = []
for num in refined_price:
    price.append(num.replace(" ", "").replace(".", "").replace(",", "."))
print(price)  
```

```python
final_price_list = []
for strings in price:
    final_price_list.append(float(strings))
print(final_price_list) 
```

```python
price_df = pd.DataFrame(final_price_list)
```

```python
price_df.describe()
```

```python
rough_size = []
for characters in rough_list:
    rough_size.append(characters[-6:])
print(rough_size)
```

```python
final_size = []
for sizes in rough_size:
    final_size.append(sizes.replace(";", "").replace(" ", ""))
print(final_size)
```

```python
final_size1 = []
for stringss in final_size:
    final_size1.append(float(stringss))
print(final_size1)
```

```python
size_df = pd.DataFrame(final_size1)
```

```python
size_df.head()
```

```python
real_estate_df = pd.concat([price_df, size_df], axis=1)
```

```python
real_estate_df.head()
```

```python
real_estate_df.columns = ['Price', 'Size(m)']
```

```python
real_estate_df.head(10)
```

```python
real_estate_df.Price / real_estate_df["Size(m)"]
```

```python
real_estate_df["Price_per_m"] = round(real_estate_df["Price"] / real_estate_df["Size(m)"], 2)
```

```python
real_estate_df
```

```python
real_estate_df['Price'].value_counts()
```

```python
real_estate_df2 = real_estate_df.groupby(["Price", "Size(m)"]).mean()
real_estate_df2 = real_estate_df2.reset_index()
real_estate_df2.shape
```

```python
real_estate_df.to_csv('data_extracted', index = False)
```
