```python
import pandas as pd
```

```python
df = pd.read_csv("full_list.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
rough_list = []
for i in df["0"]:
    rough_list.append(i)
print(rough_list)
    
    
```

```python
refined_price = []
for p in rough_list:
    refined_price.append(p.rpartition(";")[0])
print(refined_price)
    

    
#    s = '107S33M15H'    
#prefix = s.rpartition('M')[0]
```

```python
price = []
for num in refined_price:
    price.append(num.replace(" ", ""))
print(price)    
```

```python
price1 = []
for f in price:
    price1.append(f.replace(".", "").replace(",", "."))
print(price1)
```

```python
final_price_list = []
for we in price1:
    final_price_list.append(float(we))
#print(final_price_list)    
```

```python
price_df = pd.DataFrame(final_price_list)
```

```python
price_df.describe()
```

```python
rough_size = []
for u in rough_list:
    rough_size.append(u[-6:])
print(rough_size)
```

```python
final_size = []
for sizes in rough_size:
    final_size.append(sizes.replace(";", "").replace(" ", ""))
print(final_size)
```

```python
final_size1 = []
for j in final_size:
    final_size1.append(float(j))
print(final_size1)
```

```python
size_df = pd.DataFrame(final_size1)
```

```python
real_estate_df = pd.concat([price_df, size_df], axis=1)
```

```python
real_estate_df.head()
```

```python
real_estate_df.columns = ['Price', 'Size(m)']
```
